```python
fruits = np.load('fruits_300.npy')
```

```python
fruits.shape
```

```python
fruits_2d = fruits.reshape(-1, 100*100)
fruits_2d.shape
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, random_state=42)
km.fit(fruits_2d)
```

```python
km.labels_
```

```python
np.unique(km.labels_, return_counts=True)
```

```python
km.labels_ == 2
```

```python
fruits[km.labels_ == 2].shape
```

```python
fruits[km.labels_ == 0].shape
```

```python
fruits[km.labels_ == 1].shape
```

```python
def draw_fruits(arr_3d):
  '''
  arr_3d.shape: (sample #, image's width, image's height)
  '''
  n = len(arr_3d)
  rows = int(np.ceil(n/10))
  cols = n if rows < 2 else 10
  fig, axes = plt.subplots(rows, cols, squeeze=False)
  for i in range(rows):
    for j in range(cols):
      if i*10 + j < n:
        axes[i, j].imshow(arr_3d[i*10 + j], cmap='gray_r')
      axes[i, j].axis('off')
  plt.show()
```

```python
draw_fruits(fruits[km.labels_ == 2])
```

```python
km.cluster_centers_.shape
```

```python
draw_fruits(km.cluster_centers_.reshape(-1, 100, 100))
```

```python
km.transform(fruits_2d[29:30])
```

```python
km.predict(fruits_2d[29:30])
```

```python
fruits_2d[29:30].shape
```

```python
fruits_2d[29:30].reshape(1, 100, 100).shape
```

```python
draw_fruits(fruits_2d[29:30].reshape(1, 100, 100))
```

```python
inertia = []
for k in range(2, 7):
  km = KMeans(n_clusters=k, random_state=42)
  km.fit(fruits_2d)
  inertia.append(km.inertia_)

plt.plot(range(2, 7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```
